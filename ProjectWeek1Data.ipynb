Research info:
Gross domestic spending on R&D is defined as the total expenditure (current and capital) on R&D carried out by all resident companies, research institutes, university and government laboratories, etc., in a country. It includes R&D funded from abroad, but excludes domestic funds for R&D performed outside the domestic economy. This indicator is measured in USD constant prices using 2010 base year and Purchasing Power Parities (PPPs) and as percentage of GDP

```
import pandas as pd
tax = 'https://raw.githubusercontent.com/michhottinger/healthcare_spending/master/taxesPercentGDP.csv?token=AEILDQWF2Q3KOOHWDZNLCRK5V4VYK'
df_tax = pd.read_csv(tax)
df_tax.head()
```

```
rd = 'https://raw.githubusercontent.com/michhottinger/healthcare_spending/master/Gross%20domestic%20spending%20on%20R%26D.csv?token=AEILDQRPTFRSZY7HHQ3VYXK5V4TXG'
health = 'https://raw.githubusercontent.com/michhottinger/healthcare_spending/master/fullhealthcarepercapita.csv?token=AEILDQX6NLWM55Q3HJP73GK5V4T5M'
df_rd = pd.read_csv(rd)
df_health = pd.read_csv(health)
df_health.head(5)
```

```
df_tax.shape
```

```
df_rd.sort_values('LOCATION')
```

```
#Clean the healthcare spending dataset
#by selecting which columns to keep I can keep this dataset small for combining later
#select 2017 year (this is the latest year available)
#selct the percent GDP measurement to compare across datatables
#set index to location and rename 'value' because all dataframes use this column name
df_h = df_health[(df_health.TIME==2017) & (df_health.SUBJECT=='TOT') & (df_health.MEASURE=='PC_GDP')][['LOCATION', 'Value']].set_index('LOCATION').rename(columns={'Value': 'HealthCareSpending'})
df_h
```

```
#tax dataframe is selected by location and values
#'Value' column will need to be renamed and location is set to index
#location will be used to join dataframes
df_t = df_tax[['LOCATION', 'Value']].set_index('LOCATION').rename(columns={'Value': 'TaxRevenue'})
df_t.head()
```

```
df_rd.head()
```

```
#Prepare R and D data for joining
#select columns to keep, rename 'Value' column and set index to location
df_r = df_rd[df_rd.TIME==2017][['LOCATION', 'Value']].set_index('LOCATION').rename(columns={'Value': 'RandDspending'})
df_r
```

```
#Join the R&D dataframe with the Healthcare and Tax Revenue df via chaining
#sort the healthcare spending values
#drop na and reset index so Location can be used in graphs
df = df_r.join(df_t).join(df_h).sort_values('HealthCareSpending').dropna().reset_index()
df
```

All Values are expressed in Percent GDP for year 2017

```
df.shape
```

```
#start graphing to see general relationships
#needed to import %matplotlib line in order for seaborn to work
import seaborn as sns; sns.set(style="ticks", color_codes=True)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

g = sns.pairplot(df)
g
```

```
#make a graph to show Tax Revenue compared to Healthcare spending
#iterate over locations to label them on the graph

target = df['HealthCareSpending']
features = df['TaxRevenue']

fig, ax = plt.subplots(figsize=(15, 8))
sns.scatterplot(x=target, y=features, data=df, hue=df['LOCATION'], legend=False, s=200)

#set axis and title labels
ax.set_xlabel('Healthcare Spending', fontsize=16)
ax.set_ylabel('Tax Revenue', fontsize=16)
ax.set_title('Affordable Healthcare Does Not Require A High Tax Increase', fontsize=20)

for i, txt in enumerate(df.LOCATION):
    ax.annotate(txt, (target[i] + 0.15, features[i]), size=14)

plt.show;

```

```


#plot Healthcare and R&D

target = df['HealthCareSpending']
features = df['RandDspending']

#create a pretty graph
fig, ax = plt.subplots(figsize=(20, 12))
sns.scatterplot(x=target, y=features, data=df, hue=df['LOCATION'], legend=False, s=200)

# Set the x and y axes labels
ax.set_xlabel("Healthcare Spending", fontsize=16)
ax.set_ylabel("R & D Spending", fontsize=16)
ax.set_title('US Spending on R&D Cannot Explain the Healthcare Spending Issue', fontsize=20, pad=5)

#draw a line
plt.plot([2.787981, 18 ],[2.787981, 2.787981], 'k-', lw=4)

for i, txt in enumerate(df['LOCATION']):
    ax.annotate(txt, (target[i] - 0.1, features[i] - 0.1), size=12)

plt.show;
```

```
#see a basic graph of country and healcare spending
#USA has an outlier datapoint--basic overview
target = df['LOCATION']
features = df['HealthCareSpending']
fig, ax = plt.subplots(figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
sns.scatterplot(x=target, y=features, data=df)

plt.show;
```

```
#add features to make the above visualization nice

from matplotlib.pyplot import figure

#set feature(loc) and target(HCspending)
y = df['HealthCareSpending']
z = df['LOCATION']


#graph details with custom hex colors
fig, ax = plt.subplots(figsize=(15,8), dpi=70, facecolor='w', edgecolor='k')
ax.bar(z, y, color= '#16A085',  edgecolor='black')


# Set the x and y axes labels
ax.set_xlabel("Country", fontsize=16)
ax.set_ylabel("Percent GDP", fontsize=16)
ax.set_title('USA Healthcare Spending Soars', fontsize=20, pad=5)

# Fix the x an y axis tick marks and grid
#plt.autoscale(enable=True, axis='y')
#plt.autoscale(enable=True, axis='z')
ax.set_yticklabels(range(0,20,2))
ax.set_yticks(range(0,20,2));
```

```
#graph all featuers on one graph with target as %GDP
y = df['HealthCareSpending']
r = df['TaxRevenue']
z = df['RandDspending']
n = df['LOCATION']
x = np.arange(10)
fig = plt.figure()
fig, ax = plt.subplots(figsize=(25, 6), dpi=80, facecolor='w', edgecolor='k')

#break up lines so they can features
ax.plot(n, r, 'bs', label='Tax revenue')
ax.plot(n, y, 'ro', label='Healthcare Spending')
ax.plot(n, z, 'g^', label='R&D Spending')
ax.set_ylabel('Total, % of GDP, 2017', fontsize=16)
plt.title('USA Healthcare Spending Jumps in Comparision to Other Features',fontsize=20)

#add grid lines on x axis
plt.grid(axis='x', color='k', linestyle='-', linewidth=0.2)

#Place legend off to the side
chartBox = ax.get_position()
ax.set_position([chartBox.x0, chartBox.y0, chartBox.width*0.6, chartBox.height])
ax.legend(loc='upper center', bbox_to_anchor=(1.1, 0.8), shadow=True, ncol=1)
plt.show();
```
